```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.randint(1,33, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df2 = pd.DataFrame(np.random.randint(34,40, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df3 = pd.DataFrame(np.random.randint(41,50, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df4 = pd.DataFrame(np.random.randint(51,60, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df5 = pd.DataFrame(np.random.randint(61,70, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df6 = pd.DataFrame(np.random.randint(71,80, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df7 = pd.DataFrame(np.random.randint(81,100, size=(10000,5)), 
                  columns=['s1','s2','s3','s4','s5'])

df = pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
#first way
#df['obtained'] = df.loc[:,'s1':'s5'].apply(lambda x:np.sum(x), axis=1)
#df['obtained'] = df.loc[:,'s1':'s5'].apply(sum, axis=1)
# second way
#df['obtained'] = df.s1 + df.s2 + df.s3 + df.s4+ df.s5

def cal_sum(*x):
    return np.sum(x)
df['obtained'] = df.loc[:,'s1':'s5'].apply(lambda x: cal_sum(x), axis=1)
df.head()
```

```python
df['Total'] = 500
df.head()
```

```python
df['percentage'] = df.obtained / df.Total * 100
df.head()
```

# Calculate Grade with Rule Based programming
### Symbolic A.I (Classical Programming)
### Traditional Programming

```python
# Define Rule (Always defined by Human(programmer))
# Software
def grade(x):
    if x >= 80:
        return "A+"
    elif x >= 70:
        return "A"
    elif x >= 60:
        return "B"
    elif x >= 50:
        return "C"
    elif x >= 40:
        return "D"
    elif x>= 33:
        return "E"
    else:
        return "Fail"
    
df['Grade'] = df.percentage.apply(grade)
df.head()
```

```python
df.Grade.value_counts()
```

```python
df.corr().plot(kind='bar')
```

```python
#dic ={v:k for k,v in enumerate(df.Grade.unique())}
dic = {
    "A+" : 7,
    "A" : 6,
    "B" : 5,
    "C" : 4,
    "D" : 3,
    "E" : 2,
    "Fail" : 1
}
dic
```

```python
df['Grade1'] = df.Grade.map(dic)
df.head()
```

```python
df.corr().plot(kind='bar')
```

```python
df1 = df[['s1', 's2', 's3', 's4', 's5', 'Grade']]
df1.head()
```

#### apply one hot - encoding on Grade column

```python
pd.get_dummies(df1.Grade).head()
```

```python
X = df1.loc[:,'s1':'s5']
X.head()
```

```python
Y = df1.Grade
Y = pd.get_dummies(Y) # apply one-hot-encodoing on Grade
Y.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
# first-layer input_shape (no of columns in X)
# last-layer output no of labels in Y(unique values)
# last-layer activation function depend on your problem (4.6)


model = Sequential()
model.add(Dense(5, activation='relu',input_shape=(5,)))
model.add(Dense(7, activation='softmax'))
```

```python
model.summary()
```

```python
X
```

```python
# Optimizer update -> weights (w,b) based on loss function
model.compile(optimizer=SGD(lr=0.8),
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
model.fit(X,Y, epochs=8)
```

```python
df1['predict']=Y.columns.take(np.argmax(model.predict(df1.loc[:,'s1':'s5']),axis=1))

df1.head()
```
